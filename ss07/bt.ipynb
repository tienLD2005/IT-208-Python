```python
#Bai 2
import pandas as pd

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, 21, 22, 20]
score = [8.5, 7.0, 6.5, 9.0]

df = pd.DataFrame({
    "name": name,
    "age": age,
    "score": score
})

print("DataFrame filtered by loc[]:")
print(df.loc[df["score"] >= 8])

print("\nDataFrame filtered by iloc[]:")
print(df.iloc[:, [1]])

print("\nDataFrame filtered by query():")
print(df.query("score > 7 and age < 23"))
```

```python
#Bai 3
import pandas as pd

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, None, 22, 20]
score = [8.5, None, 6.5, 9.0]

df = pd.DataFrame({
    "name": name,
    "age": age,
    "score": score
})

print("Dữ liệu thiếu kiểm tra bằng isna():")
print(df.isna())

mean_age = df["age"].mean()
mean_score = df["score"].mean()

df_fill = df.copy()
df_fill["age"] = df_fill["age"].fillna(mean_age)
df_fill["score"] = df_fill["score"].fillna(mean_score)

print("\nDataFrame sau khi thay thế NaN bằng fillna():")
print(df_fill)

df_drop = df_fill.dropna()

print("\nDataFrame sau khi loại bỏ NaN bằng dropna():")
print(df_drop)
```

```python
#Bai 4
import pandas as pd

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, 21, 22, 20]
score = [8.5, 7.0, 6.5, 9.0]

df = pd.DataFrame({
    "name": name,
    "age": age,
    "score": score
})

df_apply = df.copy()
df_apply["age_group"] = df_apply["age"].apply(lambda x: "Young" if x <= 20 else "Adult")
print("DataFrame sau khi áp dụng apply():")
print(df_apply)

df_map = df.copy()
df_map["score_level"] = df_map["score"].map({
    8.5: "Excellent",
    7.0: "Good",
    6.5: "Average",
    9.0: "Excellent"
})
print("\nDataFrame sau khi map():")
print(df_map)

df_astype = df.copy()
df_astype["score"] = df_astype["score"].astype(int)
print("\nDataFrame sau khi astype():")
print(df_astype)

df_replace = df.copy()
df_replace["name"] = df_replace["name"].replace("Bình", "Bình (Updated)")
print("\nDataFrame sau khi replace():")
print(df_replace)
```

```python
#Bai 5
import pandas as pd

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, 21, 22, 20]
score = [8.5, 7.0, 6.5, 9.0]

df = pd.DataFrame({
    "name": name,
    "age": age,
    "score": score
})

grouped = df.groupby("age").agg(
    mean_score=("score", "mean"),
    min_score=("score", "min"),
    max_score=("score", "max")
).reset_index()

print("Groupby age:")
print(grouped)
```

```python
#Bai 6
import pandas as pd

df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['An', 'Bình', 'Châu']})
df2 = pd.DataFrame({'id': [1, 2, 3], 'score': [8.5, 7.0, 6.5]})

merged = pd.merge(df1, df2, on='id')
print("Merged DataFrame:")
print(merged)

df_age = pd.DataFrame({'age': [20, 21, 22]})
concatenated = pd.concat([merged, df_age], axis=1)
print("\nConcatenated DataFrame:")
print(concatenated)
```

```python
#Bai 7
import pandas as pd

date = ['2023-01-01', '2023-02-01', '2023-03-01']
name = ['An', 'Bình', 'Châu']

df = pd.DataFrame({
    "date": date,
    "name": name
})

df["date"] = pd.to_datetime(df["date"])
print("Date converted to datetime:")
print(df)

print("\nExtracted Year:")
print(df["date"].dt.year)

df["name"] = df["name"].astype("category")
print("\nCategory Type for name:")
print(df["name"])
```

```python
#Bai 8
import os
import json
import pandas as pd

log_folder = "logs/"
results = []

for filename in os.listdir(log_folder):
    if filename.endswith(".json"):
        filepath = os.path.join(log_folder, filename)
        date_str = filename.replace("log_", "").replace(".json", "")
        with open(filepath, "r", encoding="utf-8") as f:
            data = json.load(f)

        duration = data.get("duration", 0)

        results.append({
            "date": date_str,
            "total_duration": duration
        })
df = pd.DataFrame(results)

print(df)

df.to_csv("report.csv", index=False)
print("\nSaved to report.csv")
```

```python
#bai 1
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
scores = np.random.normal(7, 1.5, 200)
plt.figure(figsize=(10, 6))
sns.histplot(scores, bins=20, kde=True)

plt.title("Phân phối điểm thi – Histogram + KDE", fontsize=16)
plt.xlabel("Điểm", fontsize=13)
plt.ylabel("Tần suất / Mật độ", fontsize=13)

plt.show()
```

```python
#bai 2
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

subjects = ["Toán", "Lý", "Hóa"]
scores = np.concatenate([
    np.random.normal(7.5, 1.2, 200),
    np.random.normal(6.8, 1.3, 200),
    np.random.normal(6.2, 1.1, 200)
])

df = pd.DataFrame({
    "subject": np.repeat(subjects, 200),
    "score": scores
})

sns.displot(df, x="score", hue="subject", kind="kde", height=5, aspect=1.3)
plt.show()
```

```python
#bai 3
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

hours = np.arange(1, 11)
noise = np.random.normal(0, 1, len(hours))
score = hours * 1.2 + noise

sns.scatterplot(x=hours, y=score)
plt.xlabel("Số giờ học")
plt.ylabel("Điểm")
plt.title("Mối quan hệ giữa giờ học và điểm số")
plt.show()
```

```python
#Bai 4
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

classes = np.repeat(["A", "B", "C"], 50)
scores = np.concatenate([
    np.random.normal(8.0, 0.8, 50),
    np.random.normal(7.0, 0.7, 50),
    np.random.normal(7.6, 0.6, 50)
])

df = pd.DataFrame({"class": classes, "score": scores})

sns.barplot(data=df, x="class", y="score")
plt.title("Điểm trung bình theo từng lớp")
plt.xlabel("Lớp")
plt.ylabel("Điểm trung bình")
plt.show()
```

```python
#bai 5
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
n = 80
majors = np.repeat(["CNTT", "Kinh Tế", "Ngôn Ngữ"], n)
scores_cntt = np.random.normal(8.0, 0.9, n)
scores_kinh = np.random.normal(7.0, 1.2, n)
scores_ngon = np.random.normal(6.7, 0.7, n)
scores = np.concatenate([scores_cntt, scores_kinh, scores_ngon])
scores = np.clip(scores, 0, 10)
df = pd.DataFrame({"major": majors, "score": scores})

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.boxplot(data=df, x="major", y="score", ax=axes[0])
axes[0].set_title("Boxplot phân phối điểm theo ngành")
axes[0].set_xlabel("Ngành")
axes[0].set_ylabel("Điểm")

sns.violinplot(data=df, x="major", y="score", inner="quartile", ax=axes[1])
axes[1].set_title("Violinplot phân phối điểm theo ngành")
axes[1].set_xlabel("Ngành")
axes[1].set_ylabel("Điểm")

plt.tight_layout()
plt.show()
```

```python
#Bai 6
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

day = np.arange(1, 31)
trend = day * 1.5
noise = np.random.normal(0, 10, len(day))
sales = trend + noise

df = pd.DataFrame({"day": day, "sales": sales})

plt.figure(figsize=(12, 5))
sns.lineplot(data=df, x="day", y="sales")
plt.title("Doanh thu theo ngày (Lineplot)")
plt.xlabel("Ngày")
plt.ylabel("Doanh thu")
plt.show()

plt.figure(figsize=(12, 5))
sns.regplot(data=df, x="day", y="sales")
plt.title("Xu hướng doanh thu (Regplot)")
plt.xlabel("Ngày")
plt.ylabel("Doanh thu")
plt.show()
```

```python
#Baii 7
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

n = 500
hours = np.random.uniform(1, 10, n)
noise = np.random.normal(0, 1.5, n)
score = hours * 1.2 + noise
score = np.clip(score, 0, 10)
major = np.random.choice(["A", "B", "C"], size=n, p=[0.33, 0.34, 0.33])

df = pd.DataFrame({
    "hours": hours,
    "score": score,
    "major": major
})

plt.figure(figsize=(12, 5))
sns.histplot(df["score"], bins=20, kde=True)
plt.title("Phân phối điểm thi")
plt.xlabel("Điểm")
plt.ylabel("Tần suất")
plt.show()

plt.figure(figsize=(12, 5))
sns.scatterplot(data=df, x="hours", y="score", hue="major")
plt.title("Mối quan hệ giữa giờ học và điểm")
plt.xlabel("Giờ học")
plt.ylabel("Điểm")
plt.show()

plt.figure(figsize=(12, 5))
sns.boxplot(data=df, x="major", y="score")
plt.title("So sánh điểm theo nhóm ngành")
plt.xlabel("Ngành")
plt.ylabel("Điểm")
plt.show()
```

```python
#Bai 8
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

months = np.arange(1, 13)
categories = ["Shoes", "Clothes", "Bags"]

data = []
for cat in categories:
    base = np.random.uniform(80, 120)
    trend = np.linspace(0, np.random.uniform(-10, 15), 12)
    noise = np.random.normal(0, 8, 12)
    revenue = base + trend + noise
    data.append(pd.DataFrame({
        "month": months,
        "revenue": revenue,
        "category": cat
    }))

df = pd.concat(data, ignore_index=True)

plt.figure(figsize=(12, 5))
sns.lineplot(data=df, x="month", y="revenue", hue="category", marker="o")
plt.title("Doanh thu theo tháng (Lineplot)")
plt.xlabel("Tháng")
plt.ylabel("Doanh thu")
plt.show()

plt.figure(figsize=(12, 5))
sns.violinplot(data=df, x="category", y="revenue", inner="quartile")
plt.title("Phân phối doanh thu theo nhóm (Violinplot)")
plt.xlabel("Nhóm sản phẩm")
plt.ylabel("Doanh thu")
plt.show()

plt.figure(figsize=(12, 5))
sns.regplot(data=df, x="month", y="revenue", scatter=False, ci=None)
sns.scatterplot(data=df, x="month", y="revenue", hue="category")
plt.title("Xu hướng doanh thu chung (Regplot)")
plt.xlabel("Tháng")
plt.ylabel("Doanh thu")
plt.show()
```

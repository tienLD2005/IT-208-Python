```python
#Bai 1
try:
    with open("file.txt", "r", encoding="utf-8") as f:
        content = f.read()
        print("=== Nội dung file hiện tại ===")
        print(content)
except FileNotFoundError:
    print("File chưa tồn tại. Sẽ tạo file mới.")
    content = ""

new_line = input("Nhập dòng văn bản muốn thêm: ")

with open("file.txt", "a", encoding="utf-8") as f:
    f.write("\n" + new_line)

print("Đã ghi nội dung mới vào file.txt")
```

```python
#Bai 2
import csv

students = []
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    next(reader)  # bỏ qua dòng header

    for row in reader:
        name, age, grade = row
        if float(grade) >= 7:
            students.append(row)

print("Danh sách học sinh có grade >= 7:")
for s in students:
    print(s)

with open("high_grade_students.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerows(students)

```

```python
#Bai 3
import json

with open("user_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

for user in data:
    if user["id"] == 3:
        print(user["name"], user["age"])

new_user = {
    "id": int(input("Nhập id: ")),
    "name": input("Nhập name: "),
    "age": int(input("Nhập age: "))
}

data.append(new_user)

with open("user_data.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print("Đã thêm user mới vào file user_data.json")
```

```python
#Bai 6
import json

with open("inventory.json", "r", encoding="utf-8") as f:
    data = json.load(f)

new_product = {
    "product_id": int(input("Nhập product_id: ")),
    "name": input("Nhập tên sản phẩm: "),
    "quantity": int(input("Nhập số lượng: ")),
    "price": float(input("Nhập giá: "))
}

data.append(new_product)

with open("inventory.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print("Đã thêm sản phẩm mới vào inventory.json")
```

```python
#Bai 8
from datetime import datetime

today = datetime.today().date()

with open("log.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

for line in lines:
    timestamp_str = line.split(" - ")[0]
    timestamp = datetime.strptime(timestamp_str, "%Y-%m-%d %H:%M:%S")
    if timestamp.date() == today:
        print(line.strip())
```

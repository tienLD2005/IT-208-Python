```python
#Bai 5
import numpy as np
import time

arr = np.random.randint(1, 21, size=10)
print("Mảng ban đầu:", arr)

squared = arr ** 2    
sqrt_arr = np.sqrt(arr)  

print("\nBình phương (vectorized):", squared)
print("Căn bậc hai (vectorized):", sqrt_arr)

add_result = np.add(arr, 5)        
mul_result = np.multiply(arr, 3)  
sqrt_ufunc = np.sqrt(arr)        

print("\nKết quả dùng ufunc:")
print("np.add(arr, 5)       :", add_result)
print("np.multiply(arr, 3)  :", mul_result)
print("np.sqrt(arr)         :", sqrt_ufunc)

arr2D = np.random.randint(1, 100, size=(1000, 1000))

start_vec = time.time()
res_vec = arr2D * 2 + 5
end_vec = time.time()

start_loop = time.time()
res_loop = np.zeros_like(arr2D)
for i in range(arr2D.shape[0]):
    for j in range(arr2D.shape[1]):
        res_loop[i, j] = arr2D[i, j] * 2 + 5
end_loop = time.time()

print("\n--- So sánh tốc độ ---")
print("Vectorized time:", end_vec - start_vec, "seconds")
print("Loop time      :", end_loop - start_loop, "seconds")
```

```python
#Bai 6
import numpy as np

arr = np.array([
    [3, 7, 4, 5],
    [3, 5, 2, 4],
    [6, 3, 9, 7],
    [4, 8, 3, 5]
])

sum_col = np.sum(arr, axis=0)
mean_col = np.mean(arr, axis=0)
std_col = np.std(arr, axis=0)
sum_row = np.sum(arr, axis=1)
mean_row = np.mean(arr, axis=1)
std_row = np.std(arr, axis=1)

print("\nTổng theo cột (axis=0):", sum_col)
print("Trung bình theo cột (axis=0):", mean_col)
print("Độ lệch chuẩn theo cột (axis=0):", std_col)

print("\nTổng theo hàng (axis=1):", sum_row)
print("Trung bình theo hàng (axis=1):", mean_row)
print("Độ lệch chuẩn theo hàng (axis=1):", std_row)
```

```python
#Bai 7
import numpy as np
import time
A = np.array([
    [ 8, 12, 14,  6,  4],
    [18, 13,  2,  2, 12],
    [ 6, 18,  5, 10, 10],
    [ 9,  2,  4, 12, 12],
    [ 2,  3,  2, 17,  3]
])
B = np.array([10, 20, 30, 40, 50])
add_broad = A + B
sub_broad = A - B

print("\nKết quả cộng (broadcasting):")
print(add_broad)

print("\nKết quả trừ (broadcasting):")
print(sub_broad)

sqrt_res = np.sqrt(A)
print("\nKết quả căn bậc hai (ufunc):")
print(sqrt_res)

add5_res = np.add(A, 5)
print("\nKết quả cộng 5 vào mảng (ufunc):")
print(add5_res)

start_loop = time.time()
result_loop = np.zeros_like(A)

for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        result_loop[i, j] = A[i, j] + B[j]   

end_loop = time.time()
loop_time = end_loop - start_loop

print("\nThời gian thực thi vòng lặp thủ công:", round(loop_time, 6), "giây")
```

```python
#Bai 8
import numpy as np

A = np.array([
    [6, 9, 7],
    [2, 1, 5],
    [7, 8, 3]
])
inv_A = np.linalg.inv(A)
print("\nMa trận nghịch đảo:\n")
print(inv_A)

eigen_values, eigen_vectors = np.linalg.eig(A)

print("\nGiá trị riêng của ma trận:", eigen_values)

print("\nVéc tơ riêng của ma trận:\n")
print(eigen_vectors)
```

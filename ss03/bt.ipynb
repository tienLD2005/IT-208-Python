```python
#Bai 1
def sum_two_numbers(a:int, b:int)->int:
    return a + b

def sub_two_numbers(a:int, b:int)->int:
    return a - b

a = int(input("Nhap vao so a: "))
b = int(input("Nhap vao so b: "))

print("Tong cua hai so la: ", sum_two_numbers(a,b))
print("Hieu cua hai so la: ", sub_two_numbers(a,b))
```

```python
#Bai 2
def check_is_prime(n:int)->bool:
    if n <=1:
        print(f"{n} khong phai la so nguyen to")
        return False
    if n == 2:
        print(f"{n} la so nguyen to chan duy nhat")
        return True
    if n % 2 == 0:
        print(f"{n} khong phai so nguyen to")
        return False
    i = 3
    while i * i <= n: #Tương đương i <= sqrt(n) nhưng tốc độ nhanh hơn 
        if n % i == 0:
            print(f"{n} khong phai la so nguyen to")
            return False
        i +=2
    print(f"{n} la so nguyen to")
    return True
n = int(input("Nhap vao so nguyen n: "))
check_is_prime(n)
    
```

```python
#Bai 3
def factorial(n: int)->int:
    if n >=1:
        total = 1
        for i in range(n,1,-1):
            total *= i
        print(f"Giai thua cua gia thua {n} la: {total}")
        return total
    else:
        print(f"So khong hop le")
        return False
n = int(input("Nhap vao so n: "))
dec = factorial(n) #gán giá trị mà không iiin ra sẽ giúp giảm số lần in từ return hiện tại từ 2 -> 1
```

```python
#Bai 4
def circle_area(radius: float) -> float:
    if radius <= 0:
        print("Ban kinh khong hop le!")
        return None
    return 3.14159 * radius * radius

rad = float(input("Nhap vao ban kinh: "))

area = circle_area(rad)
if area is not None:
    print(f"Dien tich hinh tron co ban kinh {rad} la: {area}")
```

```python
#Bai 5
def is_leap_year(year)->bool:
    if year % 400 == 0:
        return True
    elif year % 4 == 0:
        if year % 100 != 0:
            return True
        else:
            return False
    return False

year = int(input("Nhap vao 1 nam: "))
if is_leap_year(year):
    print(f"{year} la nam nhuan")
else:
    print(f"{year} khong phai la nam nhuan")    
```

```python
#Bai 6
def sum_digits(number:int)->int:
    total = 0
    while number > 0:
        digits = number % 10
        total += digits
        number = number // 10
    return total
num = int(input("Nhap vao 1 so bat ki: "))
print(f"Tong cac chu so cua {num} la: {sum_digits(num)}")
```

```python
#Bai 7
def reverse_number(number):
    newNumber = ""
    for i in range(len(number)-1,-1,-1):
        newNumber += number[i]
    return newNumber
num = input("Nhap vao 1 so: ")
print(f"So dao nguoc cua {num} la: {int(reverse_number(num))}")
```

```python
#Bai 8
def check_perfect_number(num: int)->bool:
    if num <= 1:
        return False
    total = 0
    for i in range(1,num):
        if num % i == 0:
            total += i
    return total == num 
    
value = int(input("Nhap vao 1 so: "))
if check_perfect_number(value):
    print(f"So {value} la so hoan hao")
else:
    print(f"So {value} khong phai la so hoan hao")
```

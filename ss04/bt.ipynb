```python
#bai tong hop
products = []  

def add_product():
    ma = input("Nhập mã sản phẩm: ")
    for p in products:
        if p[0] == ma:
            print("Mã sản phẩm đã tồn tại! Không thể thêm.")
            return
    
    ten = input("Nhập tên sản phẩm: ")
    so_luong = int(input("Nhập số lượng: "))
    gia = float(input("Nhập giá sản phẩm: "))

    new_product = (ma, ten, so_luong, gia)
    products.append(new_product)

    print("Thêm sản phẩm thành công!")

def show_products():
    if not products:
        print("Danh sách rỗng!")
        return

    print("\n===== DANH SÁCH SẢN PHẨM =====")
    for p in products:
        print(f"Mã: {p[0]}, Tên: {p[1]}, SL: {p[2]}, Giá: {p[3]}")
    print("==============================\n")

def find_product():
    ma = input("Nhập mã sản phẩm cần tìm: ")

    for p in products:
        if p[0] == ma:
            print(f"✔ Tìm thấy: Mã={p[0]}, Tên={p[1]}, SL={p[2]}, Giá={p[3]}")
            return
    
    print("Không tìm thấy sản phẩm.")

def update_product():
    ma = input("Nhập mã sản phẩm cần cập nhật: ")

    for i, p in enumerate(products):
        if p[0] == ma:
            print("Nhập thông tin mới:")
            ten = input("Tên mới: ")
            so_luong = int(input("Số lượng mới: "))
            gia = float(input("Giá mới: "))

            products[i] = (ma, ten, so_luong, gia)
            print("Cập nhật sản phẩm thành công!")
            return

    print("Không tìm thấy sản phẩm để cập nhật.")

def delete_product():
    ma = input("Nhập mã sản phẩm cần xóa: ")

    for i, p in enumerate(products):
        if p[0] == ma:
            del products[i]
            print("Xóa sản phẩm thành công!")
            return
    
    print("Không tìm thấy sản phẩm để xóa.")

def menu():
    while True:
        print("\n===== MENU QUẢN LÝ SẢN PHẨM =====")
        print("1. Thêm sản phẩm mới")
        print("2. Hiển thị danh sách sản phẩm")
        print("3. Tìm kiếm sản phẩm theo mã")
        print("4. Cập nhật thông tin sản phẩm")
        print("5. Xóa sản phẩm theo mã")
        print("0. Thoát")
        print("==================================")

        choice = input("Chọn chức năng: ")

        if choice == "1":
            add_product()
        elif choice == "2":
            show_products()
        elif choice == "3":
            find_product()
        elif choice == "4":
            update_product()
        elif choice == "5":
            delete_product()
        elif choice == "0":
            print("Thoát chương trình!")
            break
        else:
            print("Lựa chọn không hợp lệ, vui lòng nhập lại.")
menu()
```

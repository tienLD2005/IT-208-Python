```python
#Bai 3
value = int(input("Nhap vao 1 so nguyen: "))

if value % 5 == 0 and value % 3 == 0:
    print("So vua nhap chia het cho ca 3 va 5")
elif value % 3 == 0:
    print("So vua nhap chia het cho 3")
elif value % 5 == 0:
    print("So vua nhap chia het cho 5")
else:
    print("Khong chia het cho 3 hoac 5")
```

```python
#Bai 4
n = int(input("Nhap vao 1 so n: "))
if n <= 0:
    print("So khong hop le")
else:
    total = 0
    for i in range(1,n+1,1):
        total+=i
    print(f"Tong = {total}")
```

```python
#Bai 5
n = input("Nhap 1 so")
origin = n
reveredNumber = ""
for i in range(len(n)-1,-1, -1):
    reveredNumber += n[i]
if int(origin) == int(reveredNumber):
    print("So vua nhap la so doi xung")
else:
    print("So vua nhap khong phai la so doi xung")
```

```python
#Bai 6
n = int(input("Nhap vao 1 so nguyen duong"))
if n <= 0:
    print("So khong hop le")
else:
    total = 1
    for i in range(n, 1,-1):
        total *= i
    print(f"Giai thua cua {n} la {total}")    
```

```python
#Bai 7
start = int(input("Nhap vao so bat dau"))
end = int(input("Nhap vao so ket thuc"))

if start <= 1:
    print("So bat dau phai lon hon 1")
else:
    for i in range(start, end+1, 1):
        isPrime = True
        for j in range(start, i-1):
            if i % j == 0:
                isPrime = False
                break
        if isPrime:
            print(i, end=" ")

    
```

```python
#Bai 8
edge1 = int(input("Nhap canh thu nhat: "))
edge2 = int(input("Nhap canh thu hai: "))
edge3 = int(input("Nhap canh thu ba: "))

if(edge1 > 0 and edge2 > 0 and edge3 > 0):
    if(edge1 + edge2 > edge3 and edge1 + edge3 > edge2 and edge2 + edge3 > edge1):
        print("Day la 3 canh cua tam giac")
    else:
        print("Day khong phai la 3 canh cua tam giac")
else:
    print("Do dai canh phai la so duong")
```

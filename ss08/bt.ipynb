```python
#Bai 5
import pandas as pd

with open("students.csv", "w", encoding="utf-8") as f:
    f.write("id,name\n1,An\n2,Binh\n3,Chau\n")

with open("scores.csv", "w", encoding="utf-8") as f:
    f.write("id,score\n1,8\n2,9\n3,7\n")

students = pd.read_csv("students.csv")
scores = pd.read_csv("scores.csv")

df = students.merge(scores, on="id")

report = df.groupby(["id", "name"])["score"].agg(
    score_min="min",
    score_max="max",
    score_mean="mean"
).reset_index()

report.to_csv("report.csv", index=False)

print(report)
```

```python
#Bai 6
import numpy as np

a = np.array([[1], [2], [3], [4]])
b = np.array([[10, 20, 30, 40, 50]])

res = a * b
print("a * b:\n", res)

sqrt_res = np.sqrt(res)
print("sqrt:\n", sqrt_res)

log_res = np.log(res)
print("log:\n", log_res)

mean_axis1 = res.mean(axis=1)
print("mean axis=1:", mean_axis1)

min_axis1 = res.min(axis=1)
max_axis1 = res.max(axis=1)
std_axis1 = res.std(axis=1)

print("min axis=1:", min_axis1)
print("max axis=1:", max_axis1)
print("std axis=1:", std_axis1)
```

```python
#Bai 7
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "name": ["An", "Bình", "Châu", "Dũng"],
    "age": [20, np.nan, 22, 21],
    "score": [8.5, np.nan, 7.0, 9.0],
    "city": ["Hà Nội", "HCM", np.nan, "Hà Nội"]
})

missing = df.isna()
print(missing)

df["age"] = df["age"].fillna(df["age"].mean())
df["score"] = df["score"].fillna(df["score"].median())
df["city"] = df["city"].fillna("Unknown")

df["city"] = df["city"].astype("category")

def level(x):
    if x >= 8:
        return "Good"
    elif x >= 7:
        return "Average"
    return "Bad"

df["score_level"] = df["score"].apply(level)

age_mean = df["age"].mean()
age_std = df["age"].std()
df["age_normalized"] = (df["age"] - age_mean) / age_std

print(df)
```

```python
#Bai 8
from pathlib import Path
from datetime import datetime
import json
import pandas as pd

log_dir = Path("logs")
files = list(log_dir.glob("log_*.json"))

records = []

for f in files:
    date_str = f.stem.split("_")[1]
    day = datetime.strptime(date_str, "%Y-%m-%d").date()
    with open(f, "r", encoding="utf-8") as file:
        data = json.load(file)
        data["date"] = day
        records.append(data)

df = pd.DataFrame(records)

report = df.groupby("date")["duration"].sum().reset_index()

report.to_csv("daily_report.csv", index=False)

print(report)
```
